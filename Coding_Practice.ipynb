```python
cols = ['streams', 'bpm',
       'key', 'mode', 'danceability_%', 'valence_%', 'energy_%',
       'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%']
```

```python
music = dt.loc[: , cols]
```

```python
music.info()
```

```python
# change the streams' format from object to numerical
```

```python
music[~music.streams.str.isnumeric()]
```

```python
music['streams'] = pd.to_numeric(music.streams, errors = 'coerce')
```

```python
music.dropna(inplace = True)
```

```python
music.info()
```

```python
music.rename(columns = {
    'danceability_%': 'danceability',
    'valence_%': 'valence',
    'energy_%': 'energy',
    'acousticness_%': 'acousticness',
    'instrumentalness_%': 'instrumentalness',
    'liveness_%': 'liveness',
    'speechiness_%': 'speechiness'
}, inplace = True)
```

```python
music
```

```python
music.describe()
```

```python
# EDA categorical features
plt.figure(figsize = (10, 6))
ax = sns.barplot(y = music.groupby(by = 'key').agg({'streams': 'mean'}).sort_values(by = 'streams', ascending = False).index, 
            x = music.groupby(by = 'key').agg({'streams': 'mean'}).sort_values(by = 'streams', ascending = False).streams)
ax.bar_label(ax.containers[0])
plt.title('Average of Streams By Key');
```

```python
plt.figure(figsize = (10, 6))
ax = sns.barplot(x = music['key'].value_counts().values, y = music['key'].value_counts().index)
ax.bar_label(ax.containers[0])
plt.title('Count Per Key');
```

```python
plt.figure(figsize=(10, 8))
ax = sns.barplot(y = music.groupby(by = 'mode').agg({'streams': 'mean'}).sort_values(by = 'streams', ascending = False).index,
                x = music.groupby(by = 'mode').agg({'streams': 'mean'}).sort_values(by = 'streams', ascending = False).streams)
ax.bar_label(ax.containers[0])
plt.title('Average of Streams By Mode');
```

```python
plt.figure(figsize = (10, 6))
ax = sns.barplot(y = music['mode'].value_counts().index, 
                 x = music['mode'].value_counts().values)
ax.bar_label(ax.containers[0])
plt.title('Count Per Mode');
```

```python
# EDA for numerical features
num_data = music.select_dtypes(include = ['int', 'float'])
```

```python
num_data.columns
```

```python
def scatter(col):
    plt.figure(figsize = (8, 6))
    sns.scatterplot(data = num_data, x = col, y = 'streams')
    plt.title(f'{col} vs streams');
```
